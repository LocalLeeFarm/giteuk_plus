### 모듈 로드

```python
import pandas as pd
```

```python
df1 = pd.read_csv('D:\Git\giteuk_plus\Data\Rawdata/cluster.csv',encoding='cp949')
df2 = pd.read_csv('D:\Git\giteuk_plus\Data\Preprocessing/활성화지표_추가본.csv',encoding='cp949')
```

```python
p5 = df1.자치구.str.split(' ')
p5 = p5.apply(lambda x: pd.Series(x))
p5 = p5.rename(columns={1:'자치구'})
df1 = df1.drop(['자치구'], axis=1)
df1 = pd.concat([df1,p5.loc[:,'자치구']],axis=1)
```

```python
df1
```

```python
df1 = df1[['자치구', '행정동명', '재생도시']]
```

```python
df2 =df2[['자치구','행정동명','행정동코드']]
```

```python
result = pd.merge(df1,df2)[['행정동코드','재생도시']]
```

```python
result.to_csv('../Data/Preprocessing/재생도시.csv',encoding='utf-8',index=False)
```
