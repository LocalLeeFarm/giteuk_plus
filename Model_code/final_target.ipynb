## 데이터 불러오기

```python
kmeans = pd.read_csv('../Data/Result/kmeans.csv',encoding='cp949') #kmeans 타겟 클러스터 데이터
kmedoids = pd.read_csv('../Data/Result/k-medoids.csv',encoding='cp949') #kmedoids 타겟 클러스터 데이터
gmm = pd.read_csv('../Data/Result/gmm.csv',encoding='cp949') #gmm타겟 클러스터 데이터
df = pd.read_csv('../Data/Preprocessing/활성화지표_추가본.csv',encoding='cp949') #전체 데이터
re_city = pd.read_csv('../Data/Preprocessing/재생도시.csv') #재생도시 데이터
```

## 세 클러스터에서 겹치는 데이터만 추출

```python
#각 모델에서 행정동 코드만 추출
k_mean = kmeans['행정동코드'] 
k_medoids = kmedoids['행정동코드']
g_mm = gmm['행정동코드']

#set 하기 위해 값만 뽑아옴
km_val = k_mean.values
kd_val = k_medoids.values
g_val = g_mm.values

#set진행
km = set(km_val)
kd= set(kd_val)
g = set(g_val)
```

```python
final = km&kd&g #set을 이용해 겹치는 데이터 추출
```

```python
final_df = pd.DataFrame(final).rename(columns={0:'행정동코드'}) #결과물 dataframe형태로 바꾸기
```

```python
final_df
```

## 행정동 코드를 이용해 타겟 데이터의 정보 추출

```python
final_target = pd.merge(final_df,raw_data,on=['행정동코드'],how='left')
final_target_all = pd.merge(final_target,re_city,on=['행정동코드'], how='left')
```

```python
final_target_all
```

```python
final_final_target = final_target_all[final_target_all['재생도시']==0.0] #현재 재생도시 진행중인곳은 대상에서 제외
```

```python
final_final_target
```

```python
final_final_target.to_csv('../Data/Result/final_target_cluster.csv',encoding='cp949',index=False)
```
